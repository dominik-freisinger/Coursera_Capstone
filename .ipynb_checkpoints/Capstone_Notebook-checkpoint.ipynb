### Create Pandas Dataframe and read European capitals

```python
df = pd.read_csv('capitals.csv') # capitals.csv has been downloaded from https://worldpopulationreview.com/continents/capitals-of-europe/
```

```python
df.drop('name', axis = 1, inplace = True)
```

```python
# Add Foursquare Category IDs to df columns
categories_url = 'https://api.foursquare.com/v2/venues/categories'

params = dict(
client_id=CLIENT_ID,
client_secret=CLIENT_SECRET,
v=VERSION
)
cat_resp = requests.get(url=categories_url, params=params)
cat_data = json.loads(cat_resp.text)
categories = {}
for category in cat_data['response']['categories']:
    df[category['name']] = np.nan
    categories[category['id']] = category['name']
```

```python
# read values into cells
for i, row in df.iterrows():
    for key in categories:
        city = row.capital,
        params = dict(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET,
        v=VERSION,
        near = city,
        categoryId = key
        )
        resp = requests.get(url=url, params=params)
        data = json.loads(resp.text)
        try: 
            df.loc[i,categories[key]] = data['response']['totalResults']
        except: 
            df.loc[i,categories[key]] = 0
    print(i," ", end='')
```

# KMeans Clustering
### Min Max Scaled

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
k = 5
kmeans_mm = KMeans(n_clusters=k, random_state=0).fit(features_scaled_min_max)
df_minmax = df
df_minmax['Category'] = kmeans_mm.labels_
map_clusters_mm = folium.Map(location = [53.0, 9.0], zoom_start = 4 )

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

#scale population from 1,10 for radius
scaler = MinMaxScaler(feature_range = (3,20))
scaler.fit(df['Population'].values.reshape(-1,1))
radii = pd.DataFrame(scaler.transform(df['Population'].values.reshape(-1,1)))
# add markers to the map
markers_colors = []
for radius, lat, lon, poi, cluster in zip(radii[0], df['lat'], df['lon'], df['capital'], df_minmax['Category']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=radius,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters_mm)
       
map_clusters
```
